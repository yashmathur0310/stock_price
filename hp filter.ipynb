```python
import yfinance as yf
import pandas as pd 
import numpy as np 
```

```python
df=yf.download("RELIANCE.NS",interval='1d',start="2016-11-01",end="2021-11-01")
df
```

```python
from statsmodels.tsa.filters.hp_filter import hpfilter
```

```python
df_cycle,df_trend=hpfilter(df['Open'],lamb=13322500)
```

```python
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
plt.figure(figsize=(15,8))
df['Open'].plot(label='Actual Price')
df_trend.plot(label='HP filter Analysis')
df_cycle.plot(label='Residuals')
plt.legend()
```
